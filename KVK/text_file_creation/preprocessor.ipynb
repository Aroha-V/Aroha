```python
import pandas as pd
df_2025=pd.read_csv(r'C:\Users\kumar\Desktop\KVKDEV\Aroha\KVK\2025\excel data\2025_data.csv')
df_2025.head(1)
```

```python
df_2022=pd.read_csv(r'C:\Users\kumar\Desktop\KVKDEV\Aroha\KVK\2022\excel_data\2022_data.csv')
df_2022.head(1)
```

```python
data=pd.concat([df_2025,df_2022],ignore_index=True)
```

```python
dict_var={}
for index in range(len(data)):
    dict_var[index] = data.iloc[index].to_dict()
```

```python
with open('data.txt', 'w+', encoding='utf-8') as mf:
    for i in dict_var:
        template = f'''
Location:State->{dict_var[i]['state']},City->{dict_var[i]['district']}
Disease:{dict_var[i]['disease']}
Occuring Date:{dict_var[i]['start_of_outbreak']}
Reporing Date:{dict_var[i]['date_of_reporting']}
Number of Cases:{dict_var[i]['no_of_cases']}
Number of Deaths:{dict_var[i]['deaths']}
Status:{dict_var[i]['status']}
Action Taken To Mitigate the Incident:
{dict_var[i]['action']}
'''
        mf.write(template)
```
